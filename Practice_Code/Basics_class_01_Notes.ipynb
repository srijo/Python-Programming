```python
# Compiled Language - C, C++ - They use Compiler for translation
    # Compiler takes the whole code at once and translates to Low level code


# Inetrpreterd Language - Javascript - They use Interpreter for Translation
    # Interpreter executes the code line by line



# Compiled and Interpreted Language - Python & Java
    # Both Compiler and interpretr for the translation - Byte code or Object code or Portable code, p- code


```

```python
# Source code - Any code which we write that is called the source code.
# Compiled Languages

C Source Code - Compiled - .exe file - Executed - Output - MAC - (They will be hgaving compilation as the defaut steo)

# User(Windows) - .exe file - Not executable - Platform Dependent
# Developer(Windows) - C Source code - Yes - Portable



# Interpreted Languages - Scripting Langauges- Perl

Source Code - Interpreter - Machine code - Executed - Output - MAC - Line by line output

# User(Windows) - .exe file - Not executable - Platform Dependent
# Developer(Windows) - C Source code - Yes - Portable


# Compiled and Interpreted Language
Python - Compile - Byte code - Interpreted - Machine code - Output - MAC

User(Windows) - Byte code - Platform independednt
Developer - Source code - Portable

```

```python
# Compile Time Error 
    # Syntax Error, IndentationError

# Run time Error
    # NameError, ValueError, TypeError etc

# Compilation stage in Python - Will only check Compile time Error
    # If you have any compile time error in source code then byte code will not be generated
    # If you do not have any compile time error in source code then byte code will be generated
    # If you have run time Error in siurce code then byte code will be generated
    
# Interpretation stage - Will only check Run time Error
    # If you have no run time Errror then line by line execution will be done
    # If you have any run time Errror then partial execution will be done


```
